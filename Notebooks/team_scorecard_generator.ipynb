```python
### This script will take 2 excel files that are the TDX Ticket Reports
### "My Group's Closed Tickets This Month" and
### "My Group's Received Tickets this month"
### to build a score card for each unique team member in those reports.
### The script will also ask a Yes/No (1/0) for Professional Development
### and what the previous closure rate was

from google.colab import files # needed for file picking
import pandas as pd # needed to import excel to dataframes for manipulation

# First, pop-up a file picker
# NOTE: If running this notebook in colab and you get a key error,
# check the local runtime files and make sure that there aren't duplicates
spreadsheets = files.upload()

# Note that this tool requires both spreadsheets to have the exact same unique
# technicians - it may be possible this is not the default case if the
# TDX reports "My Group's Closed Tickets This Month" and "My Group's Received
# Tickets this Month" are modified

# Reports should be named 'closed.xlsx' and 'received.xlsx'
# Reports from TDX can be output in excel format, but this tool
# expects each excel file to only have one sheet in it

# Create data frames to better handle the numbers
received = pd.read_excel(spreadsheets["received.xlsx"])
closed = pd.read_excel(spreadsheets["closed.xlsx"])

man_name = input("Enter Manager's name exactly as spelt on TDX Report")

# The data needs cleaned up a bit
# Manager should be removed from the scorecard generation
# But because the manager is on the team they may end up
# With tickets
try:
  received.drop(received.loc[received["Prim Resp"] == man_name].index, inplace=True)
except:
  "Name not found or entered incorrectly"

try:
  closed.drop(received.loc[received["Prim Resp"] == man_name].index, inplace=True)
except:
  "Name not fonud or enetered incorrectly"
```

```python
# The meat and potatoes
# For each unique primary responsible technician name
# Create an entry in the "team_perf" lsit

# But we first, let's create an empty list to put each entry into

team_perf = []

for tech in pd.unique(received["Prim Resp"]):
  team_perf.append({
      "Tech": tech,
      "Tickets Received": received.value_counts("Prim Resp")[tech],
      "Tickets Closed": closed.value_counts("Prim Resp")[tech],
      "Closure Rate": closed.value_counts("Prim Resp")[tech]/received.value_counts("Prim Resp")[tech],
      "Variance": (closed.value_counts("Prim Resp")[tech]/received.value_counts("Prim Resp")[tech]) - float(input("Enter %s's Previous Closure Rate: " % tech)),
      "PD": input("Enter 0 or 1 to indicate PD for the month")
  })
```

```python
# Peak at the performance dictionary if ya want!
print(team_perf)
```

```python
# Convert to a data frame so this can be exported back out to excel

team_perf_df = pd.DataFrame(team_perf)

team_perf_df.to_excel("scorecards.xlsx")
```
